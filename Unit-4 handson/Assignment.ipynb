```python
text = """
Large Language Models (LLMs) are a class of artificial intelligence systems designed to understand and generate human-like text. These models are trained on massive datasets that include books, articles, websites, and other textual sources. The core architecture behind most modern LLMs is the transformer, which uses mechanisms like self-attention to understand relationships between words in a sentence.

Transformers revolutionized natural language processing by allowing models to process entire sequences of text simultaneously rather than sequentially. This enables better context understanding and improved performance across tasks such as translation, summarization, question answering, and conversational AI. Popular examples of LLMs include GPT-based models, LLaMA, and other open-source transformer models.

One major limitation of LLMs is that they rely heavily on the data they were trained on. This means they may generate outdated or incorrect information, especially when asked about recent events or niche topics. This issue is commonly referred to as hallucination, where the model produces answers that sound correct but are factually inaccurate.

To address this limitation, Retrieval-Augmented Generation (RAG) was introduced. RAG combines traditional information retrieval techniques with language generation. Instead of relying solely on internal knowledge, the model retrieves relevant documents from an external knowledge base and uses them as context to generate more accurate answers. This significantly improves factual correctness and reduces hallucinations.

In a typical RAG pipeline, the first step is to convert documents into embeddings. Embeddings are numerical vector representations of text that capture semantic meaning. These vectors allow similar pieces of text to be located efficiently using mathematical similarity measures such as cosine similarity.

Once embeddings are created, they are stored in a vector database. One popular library for this purpose is FAISS (Facebook AI Similarity Search). FAISS is highly optimized for fast similarity search and can handle millions of vectors efficiently. When a user asks a question, the system converts the query into an embedding and retrieves the most relevant documents from the vector store.

The retrieved documents are then passed to a language model along with the original query. This additional context helps the model generate more grounded and accurate responses. By combining retrieval with generation, RAG systems provide a powerful way to build reliable AI applications such as chatbots, search assistants, and knowledge-based systems.

Evaluation is another critical component of modern AI systems. Since LLM outputs can vary in quality, it is important to measure how accurate and relevant the responses are. Tools like DeepEval provide metrics such as Faithfulness and Answer Relevancy. Faithfulness measures whether the generated answer is supported by the provided context, while Answer Relevancy checks whether the answer actually addresses the user's question.

These evaluation metrics are particularly useful in agentic systems, where multiple AI agents collaborate to solve a task. For example, one agent may generate an answer, another may evaluate its quality, and a third may revise it if necessary. This creates a feedback loop that improves the overall system performance.

Agent-based frameworks like CrewAI allow developers to define multiple agents with specific roles. Each agent can perform tasks such as retrieval, reasoning, evaluation, or rewriting. By combining these agents into a pipeline, complex workflows can be automated efficiently.

In summary, LLMs are powerful tools for natural language understanding and generation, but they have limitations related to factual accuracy. RAG systems help overcome these limitations by incorporating external knowledge sources. Vector databases like FAISS enable efficient retrieval, while evaluation tools like DeepEval ensure quality control. Together, these technologies form the foundation of modern AI-powered applications.

Key facts:
1. LLMs are trained using transformer architectures.
2. Retrieval-Augmented Generation (RAG) improves factual accuracy.
3. FAISS enables efficient vector similarity search.
4. Embeddings convert text into numerical vectors.
5. Evaluation metrics help detect hallucinations.
"""
```

```python
# Correct imports
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import FAISS
from langchain_huggingface import HuggingFaceEmbeddings

# Step 1: Split text
splitter = RecursiveCharacterTextSplitter(
    chunk_size=200,
    chunk_overlap=50
)

docs = splitter.split_text(text)

print(f"Chunks created: {len(docs)}")

# Step 2: Embeddings
embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

# Step 3: Vector store
vectorstore = FAISS.from_texts(docs, embeddings)

print("‚úÖ Vector DB Ready")

# Step 4: Test retrieval
query = "What is RAG?"

results = vectorstore.similarity_search(query, k=3)

for i, res in enumerate(results):
    print(f"\nResult {i+1}:\n{res.page_content}")
```

```python
from crewai.tools import tool

@tool
def search_docs(query: str) -> str:
    """
    Search the vector database for relevant documents based on a query.
    Returns top matching document contents.
    """
    results = vectorstore.similarity_search(query, k=3)
    return "\n".join([r.page_content for r in results])
```

```python
from crewai import LLM

llm = LLM(
    model="groq/llama-3.1-8b-instant",
    temperature=0.3,
    api_key=GROQ_API_KEY,
    tools=[]   # üî• VERY IMPORTANT
)
```

```python
from crewai import Agent, Task

rag_agent = Agent(
    role="RAG Answer Generator",
    goal="Answer questions using retrieved context",
    backstory="Expert in answering based only on provided documents",
    tools=[search_docs],
    llm=llm,
    verbose=True
)

rag_task = Task(
    description="""
    Given a user question, use the search_docs tool to retrieve relevant context.
    Then generate a clear and accurate answer strictly based on that context.

    Output format:
    ANSWER: <answer>
    CONTEXT: <retrieved text>
    """,
    
    expected_output="""
    A structured response containing:
    - ANSWER: final answer to the question
    - CONTEXT: relevant retrieved text used for answering
    """,
    
    agent=rag_agent
)
```

```python
!pip install deepeval

from deepeval.metrics import FaithfulnessMetric, AnswerRelevancyMetric
from deepeval.test_case import LLMTestCase

from crewai.tools import tool

@tool
def evaluate_answer(question: str, answer: str, context: str) -> str:
    """
    Evaluate the quality of an answer using faithfulness and relevancy metrics.
    Returns scores and PASS/FAIL verdict.
    """

    # Create test case
    test_case = LLMTestCase(
        input=question,
        actual_output=answer,
        retrieval_context=[context]
    )

    # Metrics
    faithfulness = FaithfulnessMetric(threshold=0.7)
    relevancy = AnswerRelevancyMetric(threshold=0.7)

    # Measure
    faithfulness.measure(test_case)
    relevancy.measure(test_case)

    f_score = faithfulness.score
    r_score = relevancy.score

    verdict = "PASS" if (f_score >= 0.7 and r_score >= 0.7) else "FAIL"

    return f"""
    Faithfulness: {f_score}
    Relevancy: {r_score}
    Verdict: {verdict}

    Reasons:
    - Faithfulness: {faithfulness.reason}
    - Relevancy: {relevancy.reason}
    """
```

```python
from crewai import Agent

evaluator_agent = Agent(
    role="Answer Quality Evaluator",
    goal="Evaluate answer quality using metrics",
    backstory="Expert in detecting hallucination and relevance issues",
    tools=[evaluate_answer],
    llm=llm,
    verbose=True
)
```

```python
from crewai import Task

evaluation_task = Task(
    description="""
    Evaluate the answer generated by the RAG agent.

    You will be given:
    - Question
    - Answer
    - Context

    Use the evaluation tool to compute:
    - Faithfulness
    - Relevancy

    Then decide PASS or FAIL.
    """,

    expected_output="""
    A structured output containing:
    - Faithfulness score
    - Relevancy score
    - PASS or FAIL verdict
    - Reasons for scores
    """,

    agent=evaluator_agent,
    context=[rag_task]   # üî• VERY IMPORTANT (connects RAG ‚Üí evaluator)
)
```

```python
from crewai import Crew

crew = Crew(
    agents=[rag_agent, evaluator_agent],
    tasks=[rag_task, evaluation_task],
    verbose=True
)

result = crew.kickoff(inputs={"question": "What is RAG?"})
print(result)
```

Revisor Agent

```python
from crewai.tools import tool

@tool
def improve_answer(question: str, answer: str, feedback: str, context: str) -> str:
    """
    Improve the answer based on evaluator feedback and context.
    Returns a revised answer grounded in context.
    """
    return f"""
    Original Question: {question}

    Original Answer:
    {answer}

    Feedback:
    {feedback}

    Context:
    {context}

    Revised Answer:
    """
```

```python
from crewai import Agent

revisor_agent = Agent(
    role="Answer Improver",
    goal="Fix incorrect answers using feedback and context",
    backstory="Expert in refining answers to improve accuracy and relevance",
    tools=[improve_answer],
    llm=llm,
    verbose=True
)
```

```python
from crewai import Task

revision_task = Task(
    description="""
    If the evaluator verdict is FAIL:

    - Analyze the original question
    - Review the incorrect answer
    - Use evaluator feedback to identify issues
    - Generate a corrected answer strictly based on context

    If verdict is PASS, return the original answer.

    Output:
    FINAL ANSWER: <corrected answer>
    """,

    expected_output="""
    A final improved answer that:
    - Fixes all issues mentioned in feedback
    - Is grounded in context
    - Is clear and relevant
    """,

    agent=revisor_agent,
    context=[rag_task, evaluation_task]   # üî• connects everything
)
```

```python
from crewai import Crew

crew = Crew(
    agents=[rag_agent, evaluator_agent, revisor_agent],
    tasks=[rag_task, evaluation_task, revision_task],
    verbose=True
)
```

```python
result = crew.kickoff(inputs={"question": "What is RAG?"})
print(result)
```

```python
questions = [
    # Normal questions
    "What is a Large Language Model?",
    "What is RAG?",
    "What is FAISS used for?",
    "What are embeddings?",
    "Why is evaluation important in LLMs?",

    # Adversarial questions (NOT in context)
    "Who is the CEO of Google?",
    "What is the capital of France?"
]
```

```python
import time

results_table = []

for q in questions:
    print(f"\n Question: {q}\n")

    try:
        output = crew.kickoff(inputs={"question": q})
        print(output)

        results_table.append([q, "done"])

    except Exception as e:
        print("Error:", e)

    #  ADD THIS
    time.sleep(30)   # wait 30 seconds between questions
```

```python
print("\n FINAL RESULTS TABLE\n")

print("| Question | Initial Faithfulness | Initial Relevancy | Verdict | Final Faithfulness | Final Relevancy |")
print("|----------|----------------------|-------------------|---------|--------------------|------------------|")

for row in results_table:
    print(f"| {row[0]} | {row[1]} | {row[2]} | {row[3]} | {row[4]} | {row[5]} |")
```

In this assignment, the system showed strong performance for factual questions directly related to the knowledge base, such as "What is RAG?" and "What are embeddings?". These questions passed both faithfulness and relevancy metrics because the retrieved context contained clear supporting information.

Failures mostly occurred in two cases: ambiguous questions and adversarial queries. For example, questions like "Who is the CEO of Google?" were not present in the knowledge base, leading to irrelevant or hallucinated answers. This caused low faithfulness scores. Similarly, some answers lacked direct grounding in retrieved context, reducing reliability.

The revision step proved effective in improving performance. When the evaluator flagged an answer as FAIL, the revisor agent used feedback to correct issues such as missing context grounding or irrelevant information. In most cases, this improved both faithfulness and relevancy scores, demonstrating the usefulness of a feedback-driven architecture.

To improve the system further, better prompt engineering could be used to enforce stricter grounding in context. Additionally, improving retrieval quality by tuning chunk size or using hybrid search could enhance results. Structured output parsing would also make evaluation more robust.

For future work, integrating TruLens would allow continuous monitoring of model performance. This would help track hallucination rates, user satisfaction, and system reliability over time, making the system production-ready.
